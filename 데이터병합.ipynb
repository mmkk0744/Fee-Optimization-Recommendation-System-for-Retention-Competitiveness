```python
import pandas as pd
```

```python
df_log = pd.read_csv('log_clean.csv')
df_ex = pd.read_csv('ex_clean.csv')
df_sv = pd.read_csv('sv_clean.csv') 
df_cus = pd.read_csv('cus_clean.csv')
```

```python
df_log = df_log.drop(columns = 'Unnamed: 0')
df_ex = df_ex.drop(columns = 'Unnamed: 0')
df_sv = df_sv.drop(columns = 'Unnamed: 0')
df_cus = df_cus.drop(columns = 'Unnamed: 0')

```

```python
df_sv_grouped = df_sv.groupby(['서비스번호', '서비스명', '판매자', '서비스가격', '평점', '대분류'], as_index=False)['판매수'].sum()
df_sv_grouped.to_csv('df_sv_grouped.csv')
```

```python
df_sv_grouped
```

```python
##수수료 기준 x
```

```python
df_total = pd.merge(df_log, df_cus, on='고객ID', how='inner')
df_total = pd.merge(df_total, df_ex, on='판매자', how='inner')
df_total = pd.merge(df_total, df_sv_grouped, on='서비스번호', how='inner')
df_total.isna().sum()
```

```python
df_total = df_total.drop(columns = ['서비스명_y', '판매자_y', '서비스가격_y','평점_y','대분류_y'])
```

```python
df_total.columns = df_total.columns.str.replace('_x', '')
```
