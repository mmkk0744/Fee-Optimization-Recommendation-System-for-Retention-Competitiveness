# df_log  NA값 처리

```python

# 거래 취소 여부 NA값 0으로 채우기
df_log['거래취소여부'] = df_log['거래취소여부'].fillna(0)
df_log['거래취소일자'] = df_log['거래취소일자'].fillna(0)
```

```python
idx = df_log.loc[df_log['서비스명'].isna() == True].reset_index(drop = True)
df_log.loc[df_log['서비스명'].isna() == True]
```

```python

# 거래기록 미기입값 22개 행 삭제
df_log = df_log.dropna().reset_index(drop = True)
```

```python
df_log.isna().sum()
```

```python
df_log.to_csv('log_clean.csv')
```

# df_ex 전처리

```python
k = pd.pivot_table(df_log,index = '판매자', values = '판매금액',aggfunc = 'sum').reset_index()
df_ex = pd.merge(df_ex,k,on='판매자')
df_ex['총 판매 금액'] = df_ex['판매금액']
df_ex = df_ex.drop(columns = ['판매금액'])
df_ex.to_csv('ex_clean.csv')
```

# df_sv NA값 처리

```python
df_sv.isna().sum()
```

```python
# NA값삭제
# df_ex['판매 서비스 수'] 의 총합이랑 df_sv['판매수']의 총합의 차이가 22인데
# NA 행의 판매수 가 22, 이상치라고 판단
df_sv = df_sv.dropna()
```

```python
df_sv.to_csv('sv_clean.csv')
```

# df_sv 데이터 중복 처리 후 저장

```python
df_sv_grouped = df_sv.groupby(['서비스번호', '서비스명', '판매자', '서비스가격', '평점', '대분류'], as_index=False)['판매수'].sum()
df_sv_grouped.to_csv('df_sv_grouped.csv')
```

# df_cus 결측치 처리


```python
df_cus.isna().sum()
```

```python
df_cus[['유입경로','연령','성별']] = df_cus[['유입경로','연령','성별']].fillna('기타')
```

```python
#df_cus = df_cus.dropna()
```

```python
df_cus['서비스총구매수'].sum()
```

```python
df_cus.loc[df_cus['고객ID'].isin(idx['고객ID']),'서비스총구매수'] = df_cus.loc[df_cus['고객ID'].isin(idx['고객ID']),'서비스총구매수']-1
```

```python
a = df_cus.loc[df_cus['고객ID'].isin(idx['고객ID'])]
```

```python
idx
```

```python

merged_df = pd.merge(a, idx, on='고객ID', how='inner')
merged_df['총수정요청수'] = merged_df['총수정요청수'] - merged_df['프로그램수정횟수']
merged_df['총추가결제금액'] = merged_df['총추가결제금액'] - merged_df['추가결제금액']
```

```python
a= a.reset_index(drop=True)
a['수정요청'] = merged_df['총수정요청수']
a['추가결제'] = merged_df['총추가결제금액']
```

```python
a.set_index(df_cus.loc[df_cus['고객ID'].isin(idx['고객ID'])].index )
```

```python
# df_cus의 특정 인덱스를 기준으로 a['수정']을 재인덱싱하여 할당
df_cus.loc[df_cus['고객ID'].isin(idx['고객ID']), '총수정요청수'] = a['수정요청'].set_axis(df_cus.loc[df_cus['고객ID'].isin(idx['고객ID'])].index, axis=0)
df_cus.loc[df_cus['고객ID'].isin(idx['고객ID']), '총추가결제금액'] = a['추가결제'].set_axis(df_cus.loc[df_cus['고객ID'].isin(idx['고객ID'])].index, axis=0)
```

```python
df_cus = df_cus.dropna()
```
