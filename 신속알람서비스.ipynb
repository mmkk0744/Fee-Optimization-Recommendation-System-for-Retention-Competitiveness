## 2-1) 전문가들이 외주 요청 알림을 보지 않아 거래 성사 X
1. 신속 알람 서비스를 사용하지 않는 판매자가 더 높은 거래 취소율을 보이는가

```python
df['알람B'] = np.where(df['신속 알람 서비스 사용여부'] == '미사용', 0, 1)
df['알람B']
```

```python
# 신속 알람 서비스 사용 여부에 따른 거래 취소율 계산
no_alert_cancel_rate = df[df['알람B'] == 0]['거래취소여부'].mean()
alert_cancel_rate = df[df['알람B'] == 1]['거래취소여부'].mean()
print(f"신속 알람 미사용 거래 취소율: {no_alert_cancel_rate * 100:.2f}%")
print(f"신속 알람 사용 거래 취소율: {alert_cancel_rate * 100:.2f}%")
```

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib
plt.figure(figsize=(6, 6))
plt.bar(['신속 알람 미사용', '신속 알람 사용'], [no_alert_cancel_rate, alert_cancel_rate], color=['#6084E4', '#455FA4'])
plt.ylabel('거래 취소율 (%)')
plt.title('신속 알람 서비스 사용 여부에 따른 거래 취소율 비교')
plt.show()
```

진짜 살짝 더 높음 => 그럼 이 비율에 차이가 있나? (두 모집단의 비율 차이 검정: 독립표본 검정)

```python
from scipy import stats

no_alert_cancel = df[df['알람B'] == 0]['거래취소여부']
alert_cancel = df[df['알람B'] == 1]['거래취소여부']

# 독립표본 t-검정 수행
t_stat, p_value = stats.ttest_ind(no_alert_cancel, alert_cancel)

print(f"t-통계량: {t_stat:.3f}")
print(f"p-값: {p_value:.4f}")
```

p-value가 0.05보다 작으므로 신속 알람 서비스 사용 여부에 따른 거래 취소율 차이가 있다고 주장할 수 있음.\
=> 신속 알람 서비스가 거래 성사율을 높이는 데 도움이 됨
